```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from numpy.random import randn
from pandas import Series,DataFrame

plt.plot(np.arange(10))

#print(plt.show())
```

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
plt.plot(randn(50).cumsum(),'k--')
ax1.hist(randn(100),bins=20,color='k',alpha=0.3)
ax2.scatter(np.arange(30),np.arange(30)+3*randn(30))
```

```python
fig,axes=plt.subplots(3,3,sharex=True,sharey=True)
for i in range(3):
    for j in range(3):
        axes[i,j].hist(randn(500),bins=50,color='k',alpha=0.5)
        
plt.subplots_adjust(wspace=0.1,hspace=0.1)
```

```python
from pylab import *
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
x=linspace(1,5,20)
y=x*x*x
ax.plot(x,y,'r--')
ax.plot(randn(30).cumsum()*10,color='k',linestyle='dashed',marker='o')
data=randn(30).cumsum()*10
plt.plot(data,'k--',drawstyle='steps-post',label='Default')
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum())
ticks=ax.set_xticks([0,250,500,750,1000])
#labels=ax.set_xticklabels(['one','two','three','four','five','five'],rotation=30,fontsize='small')
ax.set_title('My first matplotlib plot')
ax.set_xlabel('Stages')
ax.set_ylabel('hahah')
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum(),'k',label='one')
ax.plot(randn(1000).cumsum(),'k--',label='two')
ax.plot(randn(1000).cumsum(),'k.',label='three')
ax.legend c='best')
ax.text(1,1,'Hellow world!',family='monospace',fontsize=10)
```

```python
s=Series(np.random.randn(10).cumsum(),index=np.arange(0,100,10))
s.plot(style='ko--',kind='bar')
```

```python
df=DataFrame(np.random.randn(20,5).cumsum(0),columns=['a','b','c','d','e'],index=np.arange(0,100,5))
df.plot(use_index=True,rot=30,xlim=[0,50])
```

```python
tips=pd.read_csv('./python-project/bookdata/ch08/tips.csv')
party_counts=pd.crosstab(tips['day'],tips['size'])

party_pcts=party_counts.div(party_counts.sum(1).astype(float),axis=0)
party_pcts.plot(kind='bar',stacked=True)
```

```python
comp1=np.random.normal(0,1,size=200)
comp2=np.random.normal(10,2,size=200)
values=Series(np.concatenate([comp1,comp2]))
values.hist(bins=100,alpha=0.3,color='k',normed=True)
values.plot(kind='kde',style='k--')
```
